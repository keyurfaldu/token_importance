```python
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import GlueDataTrainingArguments, GlueDataset
from torch.utils.data.dataloader import DataLoader
from transformers.data.data_collator import default_data_collator
from torch.utils.data.sampler import SubsetRandomSampler
from token_importance_utils import AttentionBasedImportance as ABI
from textualheatmap import TextualHeatmap
```

```python
data_args = GlueDataTrainingArguments(task_name="CoLA", data_dir="/home/keyur/medhas/hf211/glue_data/CoLA/")
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')
```

```python
eval_dataset = GlueDataset(data_args, tokenizer=tokenizer, mode="dev")
```

```python
data_loader = DataLoader(eval_dataset, collate_fn=default_data_collator, sampler=SubsetRandomSampler([101, 1042]), batch_size=2)
```

```python
inputs = data_loader.__iter__().next()
```

```python
outputs = model(**inputs, output_attentions=True)
attentions=outputs[2]
```

```python
abiobj = ABI(inputs, tokenizer, attentions)
```

```python
data, tiles = abiobj.get_data_for_textual_heatmap(0, 0.65)
heatmap = TextualHeatmap(facet_titles = tiles, show_meta=False)
heatmap.set_data(data)
```

```python
data, tiles = abiobj.get_data_for_textual_heatmap(1, 0.65)
heatmap = TextualHeatmap(facet_titles = tiles, show_meta=False)
heatmap.set_data(data)
```
